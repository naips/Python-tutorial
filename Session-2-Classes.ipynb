## For loop
<font color=blue>**Unlike many languages, Python's loops do not intrinsically iterate over integers, but rather elements in sequences, like lists and tuples.<br>
This adds efficiency**</font> <br>

<font color=blue>**Syntax:**</font> <br>

for element in sequence:
> commands

```python
for i in [3, "hello", 9.5]:
    print(i)
```

<font color=blue>Notice that anything falling within the loop is indented beneath the first line, similar to if statements</font> <br>

```python
for i in (2.3, [8, 9, 10], {"city":"Santa Barbara"}):
    print(i)
```

```python
for i in ([2.3], [8, 9, 10], {"city","Santa Barbara"}):
    for j in i:
        print(j)
```

<font color=blue>The items in the iterable do not need to be the same type.<br>
In each case, the variable 'i' is given the value of the current list element, and the loop proceeds over these in sequence.<br> 
Also, one does not have to use the variable i; any variable name will do, but if an existing variable is used, its value will be overwritten by the loop.</font> <br>

<font color=blue>**It is very easy to loop over a part of a list using slicing:**</font> <br>

```python
l = [4, 6, 7, 8, 10]

for i in l[2:]:
    i = i*5
    print(i)
```

<font color=blue>**Using Python's multiple assignment capabilities, it is possible to iterate over more than one value at a time:**</font> <br>

```python
jack = [(1, 2), (3, 4), (5, 6)]

for (i,j) in jack:
    sum_ = i+j
    print(sum_)
```

```python
jill = [('indiana', 'jones'), ('Mr', 'India'), ('Cast', 'Away')]
    
for (i,j) in jill:
    sum_ = i+j
    print(sum_)
```

<font color=blue>**It is possible to iterate over sequences of numbers using the range function: <br>
 Syntax is:
    **</font> <br>


```python
for i in range(start, end, step):
    do something using i
```

```python
for i in range(0,23,2):
    print(i)
```

```python
for i in range(20,0,-2):
    print(i)
```

```python
l = [8, 10, 12]

for (i, val) in enumerate(l):
    print("The %ith element in the list is %d" % (i, val))
```

<font color=blue>**It is also possible to iterate over two lists simultaneously using the zip function:**</font> <br>

```python
l1 = [1, 2, 3]
l2 = [0, 6, 8]

for (a) in zip(l1, l2):
    print(a)
```

<font color=blue>**It simply takes two or more lists and groups them together, making tuples of corresponding list elements**</font> <br>

<font color=blue>**For loop inside the list:** <br> Python provides a convenient syntax for creating new lists from existing lists, tuples. These list comprehensions have the general form:</font> <br>

[expression for object in iterable]

```python
[i*i for i in range(5)]
```

<font color=blue>In the expression above, elements from the list created by the range function are accessed in sequence and assigned to the variable i. The new list then takes each element and squares it.</font> <br>

```python
[q**(0.5) for q in (4,9,16)]
```

```python
[character for character in "Python"]
```

```python
[k % 2 == 0 for k in range(5)]
```

<font color=blue>**It is also possible to filter items in list comprehensions using if statements:**</font> <br>

```python
[s for s in ["blue", "red", "green", "yellow"] if "l" in s]
```

```python
[s[0] for s in ["blue", "red", "green", "yellow"] if "l" in s]
```

### logical statements

<font color=blue>**While loops repeat as long as a certain boolean condition is met:**</font> <br>


```python
count = 0
while count < 5:
    print(count)
    count += 1  # This is the same as count = count + 1
```

<font color=blue>'break' is used to exit a for loop or a while loop, whereas 'continue' is used to skip the current block, and return to the "for" or "while" statement.</font> <br>

```python
number = 0

for number in range(10):
    if number == 5:
        break    # break here

    print('Number is ' + str(number))
```

<font color=blue>the variable number is initialized at 0. Then a for statement constructs the loop as long as the variable 'number' is less than 10. Within the for loop, there is an if statement that presents the condition that if the variable number is equivalent to the integer 5, then the loop will break</font> <br>

```python
number = 0

for number in range(10):
    if number == 5:
        continue    # continue here

    print('Number is ' + str(number))
```

<font color=blue>continue forces the loop to start at the next iteration while pass means "there is no code to execute here" and will continue through the remainder or the loop body</font> <br>

```python
number = 0

for number in range(10):
    if number == 5:
        pass    # pass here

    print('Number is ' + str(number))
```

</font> <br> **so actually where 'pass' is used**</font> <br>

```python
x = 10

for number in range(5,-5,-1):
    division = x/number

    print('Number is ' + str(number))
    
```

```python
x = 10
for number in range(5,-5,-1):
    try:
        division = x/number
        print('Number is ' + str(number))
    except:
        print('division by zero')
        pass
    
```

### Python Classes

<font color=blue>**Python, allows us to create new object types called classes. <br> 
We can then use these classes to create new objects of our own design. In the following example, we create a new class that describes a Car:**</font> <br>

```python
class Car:
    def __init__(self, Distance, Time = 12.0, Name = 'Maruti'):
        self.Name = Name 
        self.Distance = Distance 
        self.Time = Time 
    def Average(self):
        avg = self.Distance / self.Time
        return avg
```

```python
a = Car(Name = 'Mercedes', Distance = 30, Time = 2.0)
```

```python
b = Car(10)
```

```python
a.Name, a.Average()
```

```python
b.Name
```

```python
b.Average()
```

<font color=blue>In this example, the class statement indicates the creation of a new class called Car; <br>
All definitions for this class must be indented underneath it.
The first definition is for a special function called __init__ that is a constructor for the class, meaning this function is automatically executed by Python every time a new object of type AtomClass is created.<br> There are actually many special functions that can be defined for a class; each of these begins and ends with two underscore marks.<br>
Notice that the first argument to the __init__ function is the object self<font color=blue>
